```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# getting started with plotting
fig, ax = plt.subplots()

x=[1,2,3,4,5]
y=[2,3,6,4,10]

ax.plot(x, y)
plt.show()
```

```python
# Adding more features to the plot
fig, ax = plt.subplots()
ax.plot(x, y, marker='o', color='red', linestyle='--')

ax.set_title('Plot Title')
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

plt.show()
```

```python
# Creating multiple subplots (Small Multiples)
fig, ax = plt.subplots(2,2, sharey=True)

x1 = [1,2,3,4]
y1 = [2,3,5,1]
ax[0,0].plot(x1, y1)

x2 = [1,2,3,4]
y2 = [1,3,4,7]
ax[0,1].plot(x2, y2, color='r')

x3 = [1,2,3,4]
y3 = [2,4,3,1]
ax[1,0].plot(x3, y3)

x4 = [1,2,3,4]
y4 = [1,4,5,3]
ax[1,1].plot(x4, y4, color='r')

plt.show()
```

```python
# Create a plot and an array of axes: 2 rows, 1 column with shared y axis
fig, ax = plt.subplots(2, 1, sharey=True)

x1 = [1,2,3,4]
y1 = [2,3,5,1]
ax[0].plot(x1, y1, color='b', label='line 1')

x2 = [1,2,3,4]
y2 = [1,3,4,7]
ax[0].plot(x2, y2, color='r', label='line 2', linestyle='--')

x3 = [1,2,3,4]
y3 = [2,4,3,1]
ax[1].plot(x3, y3, color='b', label = 'line 1')

x4 = [1,2,3,4]
y4 = [1,4,5,3]
ax[1].plot(x4, y4, color='r', label='line 2', linestyle='--')
```

```python
# Plotting a chart to share the same axis with different figures
fig, ax = plt.subplots()

x1 = [1,2,3,4]
y1 = [2,3,5,1]
ax.plot(x1, y1)

ax2 = ax.twinx()
x2 = [1,2,3,4]
y2 = [1,3,4,7]
ax2.plot(x2, y2, color='r', linestyle='-.')
plt.yticks(range(0, 10, 2), color='r')

plt.show()
```
